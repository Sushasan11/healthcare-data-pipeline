```python
# Import necessary libraries 
import pandas as pd
import numpy as np
import pyspark as py
import sklearn
import plotly
import sqlalchemy
from sqlalchemy import create_engine
```

```python
file_path = '2014-2023 Medicare Fee-for-Service Geographic Variation Public Use File.csv'
df = pd.read_csv(file_path,encoding='Windows-1252', parse_dates=True, low_memory=False)
```

```python
# Remove rows where any cell contains '*' or '<11'
df = df[~df.apply(lambda row: row.astype(str).str.contains(r'\*|<11').any(), axis=1)].copy()
```

```python
# Convert selected columns to numeric data types, coercing invalid values to NaN
numeric_cols = [
    'TOT_MDCR_PYMT_PC', 'TOT_MDCR_STDZD_PYMT_PC',
    'BENE_AVG_AGE', 'BENE_FEML_PCT', 'BENE_MALE_PCT',
    'BENE_RACE_WHT_PCT', 'BENE_RACE_BLACK_PCT'
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Create a new column 'FEMALE_TO_MALE_RATIO' by dividing the percentage of female beneficiaries by the percentage of male beneficiaries
df['FEMALE_TO_MALE_RATIO'] = df['BENE_FEML_PCT'] / df['BENE_MALE_PCT']

# Create an 'AGE_GROUP' categorical column by binning 'BENE_AVG_AGE' into age ranges
df['AGE_GROUP'] = pd.cut(
    df['BENE_AVG_AGE'], 
    bins=[0, 64, 74, 84, 100],             # Define age bins
    labels=["<65", "65-74", "75-84", "85+"]  # Labels for each bin
)
```

```python
df['FEMALE_TO_MALE_RATIO']
```

```python
# Aggregate mean payment per capita by YEAR and geographic description
agg_df = (df.groupby(['YEAR', 'BENE_GEO_DESC'])['TOT_MDCR_PYMT_PC'].mean().reset_index().rename(columns={'TOT_MDCR_PYMT_PC': 'Avg_Payment_Per_Capita'}))
df = df.merge(agg_df, on=['YEAR', 'BENE_GEO_DESC'], how='left')
```

```python
agg_df['Avg_Payment_Per_Capita']
```

```python
# Save the cleaned DataFrame to a CSV file without the index column
df.to_csv("cleaned_medicare_data.csv", index=False)

# Save the aggregated DataFrame with average payments by year and geographic area to a CSV file
agg_df.to_csv("aggregated_payments_by_state_year.csv", index=False)
```

```python
# Save the original and aggregated DataFrames to PostgreSQL tables
engine = create_engine("postgresql://postgres:admin@localhost:5432/medicare_db")
df.to_sql("medicare_data", engine, index=False, if_exists="replace")
agg_df.to_sql("state_year_payments", engine, index=False, if_exists="replace")
with engine.connect() as connection:
    print("Connection successful!")
```
